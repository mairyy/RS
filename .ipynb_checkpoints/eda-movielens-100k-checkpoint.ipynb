```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

movielens_path = 'D:/Projects/RS/MovieLens-100k/ml-100k/'
```

# User Dataframe - Movie Dataframe - Rating Dataframe

```python
u_cols = ['user_id', 'age', 'gender', 'occupation', 'zip_code']
user_df = pd.read_csv(movielens_path + 'u.user', sep='|', names=u_cols, encoding='latin-1')

m_cols = ['movie_id', 'movie_title', 'release_date', 'video_release_date', 'imdb_url', 
          'unknown', 'Action', 'Adventure', 'Animation', 'Children\'s', 'Comedy', 'Crime', 
          'Documentary', 'Drama', 'Fantasy','Film-Noir', 'Horror', 'Musical', 'Mystery', 
          'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
movie_df = pd.read_csv(movielens_path + 'u.item', sep='|', names=m_cols, encoding='latin-1')

dateparse = lambda x: datetime.utcfromtimestamp(int(x)).strftime('%Y-%m-%d %H:%M:%S')
r_cols = ['user_id', 'movie_id', 'rating', 'timestamp']
rating_df = pd.read_csv(movielens_path + 'u.data', sep='\t', names=r_cols, encoding='latin-1', parse_dates=['timestamp'], 
                        date_parser=dateparse)
```

```python
genres = m_cols[5:]

genres_count = movie_df.loc[:, genres].sum().sort_values(ascending=False)

plt.figure(figsize=(12,7))
sns.barplot(x=genres_count.index, y=genres_count.values)
plt.xticks(rotation=90)
plt.xlabel('Genres')
plt.ylabel('Number of movies')

plt.show()
```

**Thống kê phim theo năm phát hành**

Định dạng release_date

```python
movie_df.release_date = pd.to_datetime(movie_df.release_date)
```

Thống kê phim theo năm phát hành

```python
plt.figure(figsize=(12,7))

year_release_count = movie_df.groupby(movie_df.release_date.dt.year).size().sort_values(ascending=False)
sns.lineplot(x = year_release_count.index, y = year_release_count.values)
plt.xlabel('Release year')
plt.ylabel('Number of moives')

plt.show()
```

**Thống kê phim theo tháng phát hành**

*Ví dụ: Năm >= 1990*

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

plt.figure(figsize=(12,7))
month_release_count = movie_df[movie_df.release_date.dt.year >= 1990].groupby(movie_df.release_date.dt.month).size()
sns.barplot(x = months, y = month_release_count.values)
plt.xlabel('Month')
plt.ylabel('Number of moives')
plt.title('Movies from 90s to now')

plt.show()
```

**Thống kê ngày phát hành phim trong tuần**

*Ví dụ: Năm >= 1900*

```python
days = ['Mon', 'Tue', 'Wed', 'Thur', 'Fri', 'Sat', 'Sun']

plt.figure(figsize=(12,7))
weekday_release_count = movie_df[movie_df.release_date.dt.year >= 1990].groupby(movie_df.release_date.dt.dayofweek).size()
sns.barplot(x = days, y = weekday_release_count.values)
plt.xlabel('Release day of week')
plt.ylabel('Number of moives')
plt.title('Movies from 90s to now')

plt.show()
```

### Thống kê dữ liệu rating
**Thống kê kích thước, kiểu dữ liệu, thông tin môi trường**

```python
rating_df.info()
```

```python
rating_df.describe()
```

Nối các dataframe

```python
rating_user_df = rating_df.join(other=user_df, how='inner', on='user_id', lsuffix='_R')
rating_user_df.sample(5)
```

```python
rating_user_movie_df = rating_user_df.join(other=movie_df, how='inner', on='movie_id', rsuffix='_M')
rating_user_movie_df.sample(5)
```

```python
rating_movie_df = rating_df.join(other=movie_df, how='inner', on='movie_id', rsuffix='_M')
rating_movie_df.sample(5)
```

**Thống kê số lượng rating theo năm**

```python
plt.figure(figsize=(12,7))
rating_year_count = rating_df.groupby(rating_df.timestamp.dt.year).size()
sns.barplot(x = rating_year_count.index, y = rating_year_count.values)
plt.xlabel('Year')
plt.show()
```

**Thống kê số lượng rating theo giới tính**

```python
plt.figure(figsize=(10,5))

gender_count = rating_user_df.groupby(['gender', 'rating']).size()
m_df = gender_count.M.sort_values()
f_df= gender_count.F.sort_values()

plt.bar(x = m_df.index, height = m_df.values, label='Male', width = 0.3, color='orange')
plt.bar(x = f_df.index, height = f_df.values, label='Female', width = 0.3, color='blue')
plt.legend()
plt.xlabel('Rating')

plt.show()
```

**Thống kê 10 bộ phim được đánh giá nhiều nhất**

Hàm vẽ biểu đồ ngang

```python
def draw_horizontal_movie(movie_titles, rating_count, title=''):
    plt.figure(figsize=(12,7))
    sns.barplot(x = rating_count, y = movie_titles, orient='h')
    plt.title(title)
    plt.xlabel('Rating count')
    plt.ylabel('Movies')
    plt.show()
```

*!0 bộ phim được đánh giá nhiều nhất*

```python
top_ten = rating_movie_df.groupby('movie_id').size().sort_values(ascending=False)[:10]
top_ten_titles = movie_df.iloc[top_ten.index].movie_title

draw_horizontal_movie(top_ten_titles, top_ten.values, 'Top 10 most rated movied')
```

*10 bộ phim được đánh giá nhiều nhất theo giới tính*

```python
top_movie_by_gender = rating_user_df.groupby(['gender', 'movie_id']).size()

for index_label in top_movie_by_gender.index.get_level_values(0).unique():
    top_ten_movie_count = top_movie_by_gender[index_label].sort_values(ascending=False)[:10]
    top_ten_movie_titles = movie_df.iloc[top_ten_movie_count.index].movie_title
    draw_horizontal_movie(top_ten_movie_titles, top_ten_movie_count.values, f'Top 10 rated movies by {index_label}')
```

*Top 10 phim được đánh giá nhiều nhất theo nhóm tuổi*

```python
top_movie_by_age = rating_user_df.groupby(['age_group', 'movie_id']).size()

for index_label in top_movie_by_age.index.get_level_values(0).unique():
    top_ten_movie_count = top_movie_by_age[index_label].sort_values(ascending=False)[:10]
    top_ten_movie_titles = movie_df.iloc[top_ten_movie_count.index].movie_title
    draw_horizontal_movie(top_ten_movie_titles, top_ten_movie_count.values, f'Top 10 rated movies by {index_label}')
```

**Thống kê 10 bộ phim được yêu thích nhất (được đánh giá cao nhất)**

*Top 10 bộ phim được yêu thích nhất*

```python
top_ten = rating_movie_df.groupby('movie_id').sum().rating.sort_values(ascending=False)[:10]
top_ten_titles = movie_df.iloc[top_ten.index].movie_title

draw_horizontal_movie(top_ten_titles, top_ten.values, 'Top 10 most favorite movied')
```

*Top 10 bộ phim được yêu thích nhất theo giới tính*

```python
top_movie_by_gender = rating_user_df.groupby(['gender', 'movie_id']).sum().rating

for index_label in top_movie_by_gender.index.get_level_values(0).unique():
    top_ten_movie_count = top_movie_by_gender[index_label].sort_values(ascending=False)[:10]
    top_ten_movie_titles = movie_df.iloc[top_ten_movie_count.index].movie_title
    draw_horizontal_movie(top_ten_movie_titles, top_ten_movie_count.values, f'Top 10 favorite movies by {index_label}')
```

*Top 10 bộ phim được yêu thích nhất theo nhóm tuổi*

```python
top_movie_by_age = rating_user_df.groupby(['age_group', 'movie_id']).sum().rating

for index_label in top_movie_by_age.index.get_level_values(0).unique():
    top_ten_movie_count = top_movie_by_age[index_label].sort_values(ascending=False)[:10]
    top_ten_movie_titles = movie_df.iloc[top_ten_movie_count.index].movie_title
    draw_horizontal_movie(top_ten_movie_titles, top_ten_movie_count.values, f'Top 10 favorite movies by {index_label}')
```
